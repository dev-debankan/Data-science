```python
sns.set_theme(style="darkgrid")  # Options: whitegrid, dark, white, ticks
```

```python
import numpy as np

x = np.array([0, 2, 3, 4, 5, 6, 7, 8, 9 ,10, 60])
y = np.sin(x)

sns.lineplot(x=x, y=y)
plt.title('Beautiful Line Plot')
plt.show()
```
