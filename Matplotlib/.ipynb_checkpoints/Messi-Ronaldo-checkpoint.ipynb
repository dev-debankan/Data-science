```python
import matplotlib.pyplot as plt 
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
messi_g_a = [77, 95, 113, 60, 67, 74, 63, 51, 54, 52, 47, 59, 47, 12, 34]
ronaldo_g_a = [65, 82, 78, 86, 78, 77, 56, 66, 50, 43, 42, 48, 17, 58, 51]
```

```python
plt.xlabel("YEAR")
plt.ylabel("G/A")
plt.title("GOAT G/A DATA PLOT")
plt.plot(years, messi_g_a,label='MESSI',color='blue')
plt.plot(years,ronaldo_g_a, label='RONALDO',color='red')
plt.grid("True")
plt.legend()
plt.show()
```
