Sorting and Ranking

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
df.sort_values('Year')
```

```python
df.sort_values(by='Year', ascending=False)  # Sorting by 'Age' in descending order
```

```python
df.sort_values(['Year','IMDb'])  # Sorting by 'year' and IMDb in descending order
```

```python
df2=df.sort_values(['Year','IMDb']).copy()
```

```python
df2
```

```python
df2.reset_index()  # Resetting the index of df2
```

```python
df2
```

```python
df2.reset_index(drop=True)  # Resetting the index of df2
```

```python
df2.reset_index(inplace=True)  # Resetting the index of df2 
#inplace is for modifying the original DataFrame without creating a new one
```

Rank


```python
df2['Rank'] = df2['IMDb'].rank()  # Adding a new column 'Rank' with the index values incremented by 1
```

```python
df2
```

```python
df2['IMDb'].rank(ascending=False)  # Ranking the 'IMDb' column in descending order
```
